```python
import pandas as pd
import mlflow
path = 'SAMPLE_DATA_SET.xlsx'

# Read and load dataset
df= pd.read_excel(path, sheet_name=[0, 1])
```

```python
X =df.get(0)
print(type(X))
X = (X.iloc[:,1:])
X.columns = [i for i in range(839)]
X = X.T
```

```python
Y = df.get(1)
Y = (Y.iloc[:,0:1])
train_data = pd.concat([X,Y], axis = 1, join='outer')
train_data.to_csv("train_data.csv",index=False, header = False)
print(train_data)
```

```python
import h2o
from h2o.automl import H2OAutoML
mlflow.set_experiment("H2O AutoML")
mlflow.set_tracking_uri(uri="http://127.0.0.1:5000")

# Start the H2O cluster (locally)
h2o.init()

train_data = h2o.import_file("train_data.csv")
print(train_data)
with mlflow.start_run(run_name="H2O AutoML"):
    # Run AutoML for 20 base models
    aml = H2OAutoML(max_models=20, seed=1)
    aml.train(y="C445", training_frame=train_data)

# View the AutoML Leaderboard
lb = aml.leaderboard
lb.head(rows=lb.nrows)
```
