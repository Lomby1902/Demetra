```python
import pycaret
pycaret.__version__
```

```python
import pandas as pd
import mlflow

path = 'SAMPLE_DATA_SET.xlsx'

# Read and load dataset
df= pd.read_excel(path, sheet_name=[0, 1])
```

```python
X =df.get(0)
print(type(X))
X = (X.iloc[:,1:])
X.columns = [i for i in range(839)]
X = X.T
```

```python
Y = df.get(1)
Y = (Y.iloc[:,0:1])
train_data = pd.concat([X,Y], axis = 1, join='outer')
train_data.to_csv("train_data.csv",index=False, header = False)
print(train_data)
```

```python
mlflow.set_tracking_uri(uri="http://127.0.0.1:5000")
# import pycaret regression and init setup
from pycaret.regression import *
s = setup(train_data, target = 'PARAM1', session_id = 123,feature_selection = True,log_experiment = True
         , experiment_name = "PyCaret")

import time
start_time = time.time()
# compare baseline models
best = compare_models()
print("--- %s seconds ---" % (time.time() - start_time))
```

```python
# plot error
plot_model(best, plot = 'error')
```

```python
# plot feature importance
plot_model(best, plot = 'feature')
```
